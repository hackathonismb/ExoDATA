```python
class HDXRepr:
    def __init__(self,pdb_code,pdb,hdx_file):
        self.pdb_code = pdb_code
        self.atom_ids = self.get_ids(pdb)
        self.stability, self.folding = self.get_hdx(hdx_file)
        ##will pick colours for all atom states
        self.stability_cols = {}
        self.folding_cols = {}
    
    def get_ids(self,pdb):
        """As I couldn't figure out what command to use to select a single residue,
        this method finds CA atom identifier for each residue."""
        ca_coords = {}

        with open(pdb) as f:
            g = f.read().splitlines()
            for line in [i for i in g if i.startswith("ATOM") and i[13:16].strip()=="CA"]:
                aa = int(line[22:28].strip())
                atom = int(line[4:13])
                ca_coords[aa] = atom
            f.close()
            
    def get_hdx(self,hdx_file):
        stability = {}
        folding = {}
        
        with open(hdx_file) as f:
            g = f.read().splitlines()
            for line in [i for i in g if not i.startswith("#")]:
                state = line.split(';')[-1].strip()
                aa = int(line.split(';')[0].strip())
                if state in ["EARLY","INTERMEDIATE","LATE"]:
                    folding[aa] = state
                else:
                    stability[aa] = state
            f.close()
            
        return stability, folding
    
    def output_folding(self):
        self.folding(self.folding)
    
    def output_stability(self):
        self.output(self.stability)
    
    def output(self,d):
        pass
```
